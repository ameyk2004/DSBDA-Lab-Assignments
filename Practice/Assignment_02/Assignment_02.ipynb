```python
rename_map = {
    "race/ethnicity" : "race_group",
    "parental level of education" : "parent_edu",
    "lunch" : "lunch_type",
    "test preparation course" : "course_status"
}

df = df.rename(columns=rename_map)
df.head()
```

```python
df["gender"] = df["gender"].str.lower()
df["gender"] = df["gender"].replace({
    "m" : "male",
    "f" : "female"
})
df.loc[df["gender"].str.contains("f", na=False), "gender"] = "female"
df["gender"] = df.gender.str.capitalize()
```

```python
df["race_group"] = df["race_group"].str.upper()
df["race_group"] = df["race_group"].str.split(" ").str[1]
```

```python
df.parent_edu = df.parent_edu.str.lower()

df.loc[df["parent_edu"].str.contains("bachelor", na=False),"parent_edu"] = "Bachelor"
df.loc[df["parent_edu"].str.contains("college", na=False),"parent_edu"] = "College"
df.loc[df["parent_edu"].str.contains("high", na=False),"parent_edu"] = "High School"
df.loc[df["parent_edu"].str.contains("master", na=False),"parent_edu"] = "Master"
df.loc[df["parent_edu"].str.contains("associate", na=False),"parent_edu"] = "Associate"
```

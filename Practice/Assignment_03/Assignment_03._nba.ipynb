```python
import pandas as pd

df = pd.read_csv('nba.csv')
df
```

```python
rename_map = {}
for col in df.columns:
    rename_map.update({col : col.lower()})
    
df = df.rename(columns=rename_map)
df.columns
```

```python
df.isna().sum()
```

```python
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 -q1
    
    lb = q1 - 1.5 * IQR
    ub = q3 + 1.5 * IQR
    
    return df[(df[col] > lb) & (df[col] < ub)]
    
```

```python
df = remove_outliers(df, "salary")
df = remove_outliers(df, "age")
df = remove_outliers(df, "number")
df = remove_outliers(df, "weight")

df.head()
```

```python
df.isna().sum()
```

```python
df["college"] = df["college"].fillna("Unkown")
```

```python
df["height"] = df["height"].astype(str)
df["height"] = df["height"].apply(lambda x: x.split("-")[0] + "." + x.split("-")[1])
df["height"] = df["height"].astype('float64')
```

```python
df.head()
```

```python
numeric_df = df.select_dtypes(include=["number"])

numeric_df.mean()
```

```python
numeric_df.mode()
```

```python
numeric_df.median()
```

```python
print("Col\tMin\tMax\tRange")
for col in numeric_df.columns:
    print(f"{col}\t{df[col].min()}\t{df[col].max()}\t{df[col].max() - (df[col].min())}")
```

```python
# Group By

# 1. Team
# 2. Age
# 3. Age group
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
team_group = df.groupby(by="team")

team_group["height"].agg(['mean', "min", "max", "std", "count"]).sort_values(by="mean", ascending=False)
```

```python
bins = [19,24,29,34,40]
labels = []
for i in range(len(bins) - 1):
    labels.append(f"{bins[i]}-{bins[i+1]}")
    
df["age_groups"] = pd.cut(df["age"], bins=bins, labels=labels, right=True)
age_range_grp = df.groupby(by="age_groups", observed=False)

age_range_grp["salary"].agg(["min", "max", "mean", "std", "count"])
```
